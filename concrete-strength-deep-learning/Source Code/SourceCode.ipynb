============================================
          Importing Libraries
============================================

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Input
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import warnings
warnings.simplefilter('ignore', FutureWarning)
```

============================================
          Load Dataset
============================================

```python
url = 'https://cocl.us/concrete_data'
data = pd.read_csv(url)
```

```python
# Drop Age column
data = data.drop('Age', axis=1)
```

```python
# Define predictors and target
X = data.drop('Strength', axis=1)
y = data['Strength']
```

============================================
          Model Function
============================================

```python
def regression_model(n_hidden=1):
    model = Sequential()
    model.add(Input(shape=(X.shape[1],)))
    
    for _ in range(n_hidden):
        model.add(Dense(10, activation='relu'))
        
    model.add(Dense(1))
    
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

============================================
                Part A)
  1 Hidden Layer - No Normalization - 50 Epochs
============================================

```python
mse_A = []
```

```python
for i in range(50):
    xtr, xts, ytr, yts = train_test_split(X, y, test_size=0.3, random_state=i)
    
    model = regression_model(n_hidden=1)
    model.fit(xtr, ytr, epochs=50, verbose=0)
    
    preds = model.predict(xts)
    mse_A.append(mean_squared_error(yts, preds))
```

```python
print("Part A Results")
print("Mean MSE:", np.mean(mse_A))
print("Std MSE:", np.std(mse_A))
```

============================================
                Part B)
  1 Hidden Layer - Normalization - 50 Epochs
============================================

```python
X_norm = (X - X.mean()) / X.std()
```

```python
mse_B = []
```

```python
for i in range(50):
    xtr, xts, ytr, yts = train_test_split(X_norm, y, test_size=0.3, random_state=i)
    
    model = regression_model(n_hidden=1)
    model.fit(xtr, ytr, epochs=50, verbose=0)
    
    preds = model.predict(xts)
    mse_B.append(mean_squared_error(yts, preds))
```

```python
print("\nPart B Results")
print("Mean MSE:", np.mean(mse_B))
print("Std MSE:", np.std(mse_B))
```

============================================
                Part C)
  1 Hidden Layer - Normalization - 100 Epochs
============================================

```python
mse_C = []
```

```python
for i in range(50):
    xtr, xts, ytr, yts = train_test_split(X_norm, y, test_size=0.3, random_state=i)
    
    model = regression_model(n_hidden=1)
    model.fit(xtr, ytr, epochs=100, verbose=0)
    
    preds = model.predict(xts)
    mse_C.append(mean_squared_error(yts, preds))
```

```python
print("\nPart C Results")
print("Mean MSE:", np.mean(mse_C))
print("Std MSE:", np.std(mse_C))
```

  3 Hidden Layers - Normalization - 100 Epochs

```python
for i in range(50):
    xtr, xts, ytr, yts = train_test_split(X_norm, y, test_size=0.3, random_state=i)
    
    model = regression_model(n_hidden=3)
    model.fit(xtr, ytr, epochs=100, verbose=0)
    
    preds = model.predict(xts)
    mse_D.append(mean_squared_error(yts, preds))
```
